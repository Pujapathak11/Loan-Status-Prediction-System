# Handle Imbalance data

```python
from imblearn.over_sampling import SMOTE
```

```python
x_rasmple, y_rasmple = SMOTE().fit_resample(x, y.values.ravel())
```

```python
print(x_rasmple.shape)
print(y_rasmple.shape)
```

```python
y.shape
```

# Train test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test = train_test_split(x_rasmple, y_rasmple, test_size = 0.2, random_state = 0)
```

```python
x_train.shape
```

```python
y_test.shape
```

# Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score, accuracy_score
```

```python
models =  [LogisticRegression, DecisionTreeClassifier, RandomForestClassifier, SVC, GaussianNB, KNeighborsClassifier]
mae_scores = []
mse_scores = []
rmse_scores = []
r2_scores = []
accuracy_scores = []

for model in models:
    classifier = model().fit(x_train, y_train)
    y_pred = classifier.predict(x_test)

    mae_scores.append(mean_absolute_error(y_test, y_pred))
    mse_scores.append(mean_squared_error(y_test, y_pred))
    rmse_scores.append(mean_squared_error(y_test, y_pred, squared = False))
    r2_scores.append(r2_score(y_test, y_pred))
    accuracy_scores.append(accuracy_score(y_test, y_pred))
```

```python
classification_metrics_df = pd.DataFrame({
    "Model" : ["LogisticRegression","DecisionTreeClassifier","RandomForestClassifier","SVC","GaussianNB","KNeighboursClassifier"],
    "Mean Absolute Error" : mae_scores,
    "Mean Squared Error" : mse_scores,
    "Root Mean Squared error" : rmse_scores,
    "R_squared(R2)" : r2_scores,
    "accuracy score" : accuracy_scores
})

classification_metrics_df.set_index('Model',inplace = True)
classification_metrics_df
```

```python
rc = RandomForestClassifier(n_estimators = 1000)
```

```python
rc.fit(x_train,y_train)
```

```python
y_pred = rc.predict(x_test)
```

```python
print("accuracy score is: ",accuracy_score(y_test,y_pred))
print("Mean Absolute error is: ",mean_absolute_error(y_test,y_pred))
print("Mean squared error is: ",mean_squared_error(y_test,y_pred))
print("R2 score is: ",r2_score(y_test,y_pred))
```
